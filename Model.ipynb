Import Required Library

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

Initialize the LLM model

```python
llm = ChatOllama(
    model="llama3.2",
    temperature=0.7
)
```

Initialize Prompt and Create a LCEL chain 

```python
prompt = PromptTemplate.from_template("""
You are an expert chef and cooking assistant. Based on the ingredients provided, suggest delicious recipes.

Ingredients: {ingrediants}
Please provide:
1. Recipe name
2. Difficulty level (Easy/Medium/Hard)
3. Cooking time
4. Step-by-step instructions
5. Pro tips

Keep your response concise and practical.
""")

chain = prompt | llm | StrOutputParser()
```

Result

```python
result = chain.invoke({"ingrediants": "chicken, rice, vegetables"})
print(result) 
```
